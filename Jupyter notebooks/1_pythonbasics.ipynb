Write a Python function named leap that takes a year as input from the user and determines whether it is a leap year or not. A leap year is divisible by 4, but not divisible by 100 unless it is also divisible by 400.

````python
def leap(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
    
user_year = int(input("Enter a year: "))

if leap(user_year):
    print(f"{user_year} is a leap year.")
else:
    print(f"{user_year} is not a leap year.")
````

9.3.2. Example: Number of Heads in 100 Tosses:

Simulation Scenario:

Simulating the number of heads in 100 tosses of a coin.

Step 1: What to Simulate:

Simulate the number of heads in 100 tosses.

Step 2: Simulating One Value:

Create a function (one_simulated_value) that simulates the number of heads in 100 tosses using np.random.choice and np.count_nonzero.

Step 3: Number of Repetitions:

Decide on the number of repetitions (20,000 in this example).

Step 4: Simulating Multiple Values:

Use a for loop to simulate the number of heads 20,000 times, collecting the results in an array (heads).
Visualize the results with a histogram.

````python
import numpy as np
import pandas as pd
import seaborn as sns

def bts(iteration):
    data = []
    for i in range(1, iteration + 1):
        outcome = np.random.choice(['Head', 'Tail'], size=500)
        heads_count = np.count_nonzero(outcome == 'Head')
        tails_count = np.count_nonzero(outcome == 'Tail')
        data.append({
            'Iteration': i,
            'Heads': heads_count,
            'Tails': tails_count
        })
    df = pd.DataFrame(data)
    return df

def main():
    iteration = int(input("Enter the number of iterations: "))
    df = bts(iteration)
    print(df.head(5))
    sns.set_palette("magma")
    sns.set_style("dark")
    sns.histplot(data=df, x='Heads')

main()
````

## 10. Sampling and distributions
Empirical Histograms, as num -> inf, hist -> uniform dist.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def dice(num):
    outcomes = np.arange(1, 7)
    probabilities = [0.15, 0.15, 0.25, 0.15, 0.15, 0.15]
    data = []
    for i in range(1, num + 1):
        dice_result = np.random.choice(outcomes, p=probabilities)
        data.append({'Roll': i, 'Result': dice_result})
    table = pd.DataFrame(data)
    return table

def vis():
    sns.set_style('dark')
    sns.histplot(x='Result',
                 data=table,
                 binwidth=1,
                 discrete=True,
                 legend = False,
                 hue = 'Result',
                 palette = 'rocket')

num = int(input("Enter the number of rolls: "))
table = dice(num)
vis()
````

## 11. Hypothesis testing

````python
# Original sample
import pandas as pd
import seaborn as sns
jury = pd.DataFrame({
    'Ethnicity': ['Asian/PI', 'Black/AA', 'Caucasian', 'Hispanic', 'Other'],
    'Eligible': [0.15, 0.18, 0.54, 0.12, 0.01],
    'Panels': [0.26, 0.08, 0.54, 0.08, 0.04]
    })
jury

melted_jury = pd.melt(jury, id_vars='Ethnicity', var_name='Category', value_name='Percentage')

sns.barplot(x='Percentage', y='Ethnicity', hue = 'Category', palette = 'mako', data=melted_jury)
````

````python
import numpy as np

# Sampling: Randomly sample 1453 times from the distribution of eligible jurors
sample_distribution = np.random.multinomial(1453, jury['Eligible']) / 1453

# Merging: Add the random sample distribution to the original jury table
jury['Random sample'] = sample_distribution

# Display both
from IPython.display import display

display(jury)

melted_jury = pd.melt(jury, id_vars=['Ethnicity'], var_name='Category', value_name='Percentage')
sns.barplot(x='Percentage', y='Ethnicity',
    hue = 'Category', palette = 'rocket',
    data = melted_jury)
````

Technical note. Random samples of prospective jurors would be selected without replacement. However, when the size of a sample is small relative to the size of the population, sampling without replacement resembles sampling with replacement; the proportions in the population don’t change much between draws. The population of eligible jurors in Alameda County is over a million, and compared to that, a sample size of about 1500 is quite small. We will therefore sample with replacement.

**TVD** measures the dist of ethinicities in eligible vs. in panels. The larger the TVD, the more different the 2 dist appear.

##### Stimulating the stat under the model

Basically we wanna see how the TVD varies across random samples -> stimulate it repeatedly under the model of random selection. 

TVD = sum(np.abs(dist1 - dist 2)) / 2
Difference -> absolute -> sum the abs column -> divide 2

````python
import numpy as np
def tvd (dist1, dist2):
    return sum(np.abs(dist1 - dist2)) / 2
````

````python
ckd = pd.read_csv('ckd.csv')
ckd.head(1)
````

````python
ckd_subset = ckd[['Hemoglobin', 'Blood Glucose Random', 'White Blood Cell Count', 'Class']]
ckd_subset = ckd_subset.rename(columns={'Hemoglobin': 'hemo', 'Blood Glucose Random': 'glucose',
                          'White Blood Cell Count': 'white', 'Class': 'class'})
ckd_subset.head(1)
````

````python
ckd_subset_stdr = ckd_subset[['hemo', 'glucose', 'white']] = (
    ckd_subset[['hemo', 'glucose', 'white']] - ckd_subset[['hemo', 'glucose', 'white']].mean()
) / ckd_subset[['hemo', 'glucose', 'white']].std()
ckd_subset_stdr['class'] = ckd_subset['class']
ckd_subset_stdr.head(1)
````

Let’s look at two columns in particular: the hemoglobin level (in the patient’s blood), and the blood glucose level (at a random time in the day; without fasting specially for the blood test).

We’ll draw a scatter plot to visualize the relation between the two variables. Blue dots are patients with CKD; gold dots are patients without CKD. What kind of medical test results seem to indicate CKD?

````python
plt.figure(figsize = (3, 3))
sns.scatterplot(data = ckd_subset_stdr,
            x = 'hemo',
            y = 'glucose',
            hue = 'class',
            palette = 'rocket')
````

You can see a very clear pattern here: points in the lower-right tend to represent people who don’t have CKD, and the rest tend to be folks with CKD.
